```python
import os 
from transformers import HfArgumentParser
from llava.train.train import train, ModelArguments, DataArguments, TrainingArguments

parser = HfArgumentParser((ModelArguments, DataArguments, TrainingArguments))

config_file_path = "configs/config.json"

# parse the config file
model_args, data_args, training_args = parser.parse_json_file(json_file=os.path.abspath(config_file_path))

train(model_args=model_args, data_args=data_args, training_args=training_args)

print("\ntraining completed!")
```
